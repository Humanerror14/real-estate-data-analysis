```python
cols_to_drop = [col for col in ["id", "Unnamed: 0"] if col in df.columns]
df.drop(cols_to_drop, axis=1, inplace=True)
df.describe()
```

```python
sns.boxplot(x="waterfront", y="price", data=df)
plt.title("Price Distribution by Waterfront")
plt.xlabel("Waterfront (1 = Yes, 0 = No)")
plt.ylabel("Price")
plt.show()
```

```python
sns.regplot(x="sqft_above", y="price", data=df)
plt.title("Correlation between sqft_above and Price")
plt.xlabel("Square Footage Above Ground")
plt.ylabel("Price")
plt.show()
```

```python
# Definisikan fitur dan target
X = df[["sqft_living"]]  # fitur harus dalam bentuk 2D array
y = df["price"]          # target

# Inisialisasi dan latih model
model = LinearRegression()
model.fit(X, y)

# Hitung R^2
r2 = model.score(X, y)
print("R^2:", r2)
```

```python
from sklearn.linear_model import Ridge
from sklearn.model_selection import train_test_split  

# Fitur yang digunakan
features = ["floors", "waterfront", "lat", "bedrooms", "sqft_basement",
            "view", "bathrooms", "sqft_living15", "sqft_above", "grade", "sqft_living"]

X = df[features]
y = df["price"]

# Bagi data ke training dan testing
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)

# Buat dan latih model Ridge
ridge_model = Ridge(alpha=0.1)
ridge_model.fit(X_train, y_train)

# Hitung R^2 pada test data
r2 = ridge_model.score(X_test, y_test)
print("R^2 on test data:", r2)
```

```python
features = ["floors", "waterfront", "lat", "bedrooms", "sqft_basement",
            "view", "bathrooms", "sqft_living15", "sqft_above", "grade", "sqft_living"]

X = df[features]
y = df["price"]

# Bagi data menjadi training dan testing
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)

# Lakukan transformasi polynomial orde 2
poly = PolynomialFeatures(degree=2, include_bias=False)
X_train_poly = poly.fit_transform(X_train)
X_test_poly = poly.transform(X_test)

# Buat dan latih model Ridge Regression
ridge_poly_model = Ridge(alpha=0.1)
ridge_poly_model.fit(X_train_poly, y_train)

# Hitung R^2 pada test data
r2 = ridge_poly_model.score(X_test_poly, y_test)
print("R^2 on test data (Polynomial + Ridge):", r2)
```
